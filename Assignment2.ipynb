```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1
# Define the membership functions for fuzzy sets A and B
def membership_function_A(x):
    if x < 30:
        return 1
    elif 30 <= x < 50:
        return (x - 30) / (50 - 30)
    elif 50 <= x < 70:
        return (x - 50) / (70 - 50)
    else:
        return 0

def membership_function_B(x):
    if x < 60:
        return 0
    elif 60 <= x < 80:
        return (x - 60) / (80 - 60)
    elif 80 <= x < 100:
        return (x - 80) / (100 - 80)
    else:
        return 1

# Define the universe of discourse (speed values)
speed_values = np.arange(10, 150)

# Calculate the membership values for fuzzy sets A and B
membership_values_A = [membership_function_A(x) for x in speed_values]
membership_values_B = [membership_function_B(x) for x in speed_values]
```

```python
# 1i) Union
membership_values_C = [max(membership_values_A[i], membership_values_B[i]) for i in range(len(speed_values))]

# 1ii) Intersection
membership_values_D = [min(membership_values_A[i], membership_values_B[i]) for i in range(len(speed_values))]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(speed_values, membership_values_A, label='Fuzzy Set A')
plt.plot(speed_values, membership_values_B, label='Fuzzy Set B')
plt.xlabel('Speed (km/hr)')
plt.ylabel('Membership Value')
plt.title('Membership Curves of Fuzzy Sets and Operations')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(speed_values, membership_values_C, label='Union (C)')
plt.plot(speed_values, membership_values_D, label='Intersection (D)')
plt.xlabel('Speed (km/hr)')
plt.ylabel('Membership Value')
plt.title('Membership Curves of Fuzzy Sets and Operations')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the membership functions for fuzzy sets C and D
def membership_function_C(x):
    if x < -10:
        return 0
    elif -10 <= x < 0:
        return (x - (-10)) / (0 - (-10))
    elif 0 <= x < 5:
        return (x - 0) / (5 - 0)
    else:
        return 0

def membership_function_D(x):
    if x < 25:
        return 0
    elif 25 <= x < 35:
        return (x - 25) / (35 - 25)
    elif 35 <= x < 40:
        return 1
    else:
        return 0


# Define the universe of discourse
U = np.arange(-20, 70)

# Calculate the membership values for fuzzy sets C and D
membership_values_C = [membership_function_C(x) for x in U]
membership_values_D = [membership_function_D(y) for y in U]

# Perform Max-Min Composition to obtain fuzzy set F
μF = np.minimum(membership_values_C, membership_values_D)
```

```python
# Plot the membership curve of fuzzy set F
plt.figure(figsize=(10, 6))
plt.plot(U, membership_values_C, label='Fuzzy Set C')
plt.plot(U, membership_values_D, label='Fuzzy Set D')
plt.plot(U, μF, label='Max-Min Composition (F)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Membership Value')
plt.title('Membership Curves of Fuzzy Sets and Max-Min Composition')
plt.legend()
plt.grid()
plt.show()
```
